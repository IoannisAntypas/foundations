```python
bad_barbers = []
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
for name in doc.find_all('tr')[2:]:
    current = {}
    Name = name.find_all(class_= 'results_text')[0].text
    if Name:
        current['Name'] = Name
        print(name.find_all(class_= 'results_text')[0].text)
    Violation = name.find_all('td')[2].text
    if Violation:
        current['Violation'] = Violation
    print(name.find_all('td')[2].text)
    complaint = name.find_all('td')[0].text
    Number = re.findall("BAR\d+",complaint,re.IGNORECASE)
    if Number:
        current['Number'] = Number
    bad_barbers.append(current)
    print(Number)
    print("-----")

```

### Save that to a CSV

```python
import pandas as pd
```

```python
df = pd.DataFrame(bad_barbers)
```
