## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Operator ID
- Operator name
- Mine name
- State
- Mine type
- Coal or metal
- Status
- Commodity

Create a new dictionary for each row.

```python
mines = []

for name in operators:
    current = {}
    Operator_ID = name.find_all('td')[0].text
    if Operator_ID:
        current['Operator_ID'] = Operator_ID
        print(Operator_ID)
    Operator_name = name.find_all('td')[1].text
    if Operator_name:
        print(Operator_name)
        current['Operator_name'] = Operator_name
    Mine_name = name.find_all('td')[2].text
    if Mine_name:
        print(Mine_name)
        current['Mine_name'] = Mine_name
    State = name.find_all('td')[3].text
    if State:
        print(State)
        current['State'] = State
    Mine_type = name.find_all('td')[4].text
    if Mine_type:
        print(Mine_type)
        current['Mine_type'] = Mine_type
    Coal_or_metal = name.find_all('td')[5].text
    if Coal_or_metal:
        print(Coal_or_metal)
        current['Coal_or_metal'] = Coal_or_metal
    Status = name.find_all('td')[6].text
    if Status:
        print(Status)
        current['Status'] = Status
    Commodity = name.find_all('td')[7].text
    if Commodity:
        print(Commodity)
        current['Commodity'] = Commodity
    print(current)
    mines.append(current)
    print("---------")
```
